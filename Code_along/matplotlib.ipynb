# Matplotlib fundemental

```python
import pandas as pd

df = pd.read_csv("02_matplotlib_bar_line/data/norway_new_car_sales_by_make.csv")
df
```

```python
df.head(10)
```

```python
df.head().T
```

```python
import duckdb

df_quantity_make = duckdb.query("""
SELECT 
    make AS brand,
    SUM(quantity) AS Quantity
    FROM df
    GROUP BY
        make
    ORDER BY
        quantity
    DESC
""").df()

df_quantity_make.head()
```

# Bar chart

```python
# using dataframes plot method which is a wrapper for matplotlib
ax = df_quantity_make.plot(kind="bar", x="Make", y="Quantity", title = "Cars sold in Norway");
```

# Matplotlib bar

```python
import matplotlib.pyplot as plt


def top_ten_bar():
    df_quantity_make_top_10 = df_quantity_make.iloc[:10]

    fig, ax = plt.subplots(figsize=(12, 6))


    ax.bar(
        x=df_quantity_make_top_10["make"],
        height=df_quantity_make_top_10["Quantity"],
        width=0.6,
    )

    ax.set(
        title="Number of cars from top 10 most sold brands in Norway 2007-2017",
        xlabel="Car brand",
        ylabel="Number of cars",
    )

    return fig, ax

fig, ax = top_ten_bar();
fig.tight_layout()
fig.savefig("figures/bars_top_10.png")
```

# Line Chart

```python
df_quantity_year = duckdb.query("""
    SELECT
        year,
    FROM df
    GROUP BY year
    ORDER BY 
        year 
""").df().set_index("Year")

df_quantity_year
```

```python
ax = df_quantity_year.plot()
ax.set(title="Yearly car sales in norway", ylabel="Number of cars");
```
